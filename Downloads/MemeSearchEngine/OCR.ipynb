```python
import ast

def getText(data):
    
    x = ast.literal_eval(data)
    textAns = ""
    try:
        regionlist = x['regions']
        for region in regionlist:
            linelist = region['lines']
            for line in linelist:
                wordlist=line['words']
                for word in wordlist:
                    textAns+=(word['text']+" ")
        textAns=textAns.strip()
        textAns=textAns.lower()
        return textAns
    except:
        return textAns
```

```python
#This code uses getText method defined in above cell. And this is used to extract and detect 
#text using MS Computer Vision API
#Change dirname and the key of the product and then this code can be used
import Image,os,csv
import httplib, urllib, base64
import time
dirname = "C:\\Users\\Abhishek\\Downloads\\memetestdata\\testmemes"
imglist = os.listdir(dirname)
imgdirectorylist = []
pathofimgtext = "memedata_test.txt"
f = open(pathofimgtext,"ab")
for img in imglist:
    try:
        Image.open(os.path.join(dirname,img))
        headers = {
            # Request headers
            'Content-Type': 'application/octet-stream',
            'Ocp-Apim-Subscription-Key': 'Enter your key',
        }

        params = urllib.urlencode({
            # Request parameters
            'language': 'unk',
            'detectOrientation ': 'true',
        })
        data = open(os.path.join(dirname,img), 'rb').read()
        conn = httplib.HTTPSConnection('api.projectoxford.ai')
        conn.request("POST", "/vision/v1.0/ocr?%s" % params, data, headers)
        response = conn.getresponse()
        data = response.read()
        print img
        textinimg = getText(data)
        if(textinimg is not ""):
            f.write(img+"#"+textinimg)
            f.write("\n")
            print "done"
        time.sleep(4)
        

    
    except IOError:
        continue

f.close()
```
